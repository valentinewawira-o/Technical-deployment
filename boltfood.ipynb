```python
import os
import pandas as pd
import shutil
from datetime import datetime

# Define paths
source_folder = "C:/Users/User/Downloads/BoltFood/Daily"  # Folder with CSV files
archive_folder = "D:/ArchiveBoltFood"  # Archive folder
output_folder = "D:/OutputBoltFood"  # Folder for combined file
output_filename = "CourierOrders_combined_output.csv"  # Fixed output filename

# Create necessary folders if they don't exist
os.makedirs(archive_folder, exist_ok=True)
os.makedirs(output_folder, exist_ok=True)

# Get all CSV files in the source folder
csv_files = [f for f in os.listdir(source_folder) if f.endswith('.csv')]

if not csv_files:
    print("No CSV files found in the source folder.")
else:
    df_list = []

    for file in csv_files:
        # Extract date from the filename (courier_orders_YYYY_MMM_DD.csv)
        try:
            # Ensure the filename follows the expected format
            filename_parts = file.replace(".csv", "").split("_")
            if len(filename_parts) != 4:
                raise ValueError(f"Unexpected filename format: {file}")

            year, month_str, day = filename_parts[1], filename_parts[2], filename_parts[3]
            
            # Convert month abbreviation (e.g., "Feb") to number
            file_date = datetime.strptime(f"{year} {month_str} {day}", "%Y %b %d").strftime("%Y-%m-%d")

        except (IndexError, ValueError) as e:
            print(f"Skipping file due to naming issue: {file}. Error: {e}")
            continue

        # Read CSV and add 'Order Date' column
        file_path = os.path.join(source_folder, file)
        try:
            df = pd.read_csv(file_path)
        except Exception as e:
            print(f"Error reading file {file}: {e}")
            continue

        df["Order Date"] = file_date  # Add extracted date
        df["Quarter"] = pd.to_datetime(df["Order Date"]).dt.to_period("Q").astype(str).str[-2:]  # Extract Q1, Q2, etc.

        df_list.append(df)

    if df_list:
        # Combine all DataFrames
        combined_df = pd.concat(df_list, ignore_index=True)

        # Generate a timestamped filename for backup
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_filename = f"Courier_combined_output_{timestamp}.csv"
        backup_path = os.path.join(archive_folder, backup_filename)

        # Define final output path
        output_path = os.path.join(output_folder, output_filename)

        # Save the combined CSV to both output and backup archive
        combined_df.to_csv(output_path, index=False)
        combined_df.to_csv(backup_path, index=False)

        print(f"Combined CSV saved to: {output_path}")
        print(f"Backup of combined CSV saved to: {backup_path}")

        # Move processed files to archive
        for file in csv_files:
            shutil.move(os.path.join(source_folder, file), os.path.join(archive_folder, file))

        print(f"All processed files moved to: {archive_folder}")

    else:
        print("No valid files were processed. Please check file formats.")
```

```python
import pandas as pd

# File paths
courier_file = "D:/OutputBoltFood/CourierOrders_combined_output.csv"
riders_file = r"C:\Users\User\Downloads\riders_info.csv"

# Load courier data
courier_df = pd.read_csv(courier_file)
courier_df.columns = courier_df.columns.str.encode('ascii', 'ignore').str.decode('ascii').str.strip()
courier_df['Phone'] = courier_df['Phone'].astype(str).str.strip()

# Load riders info
riders_df = pd.read_csv(riders_file)
riders_df.columns = riders_df.columns.str.encode('ascii', 'ignore').str.decode('ascii').str.strip()
riders_df['Phone Number'] = riders_df['Phone Number'].astype(str).str.strip()

# Create lookup dictionary from Phone Number to Riders Id
phone_to_id = dict(zip(riders_df['Phone Number'], riders_df['Riders Id']))

# Map Riders Id using Phone Number
courier_df['Riders Id'] = courier_df['Phone'].map(phone_to_id)

# Save updated file
courier_df.to_csv(courier_file, index=False)

print(" Riders Id column added based on phone number and file saved successfully.")
```
